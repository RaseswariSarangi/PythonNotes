# while loop

```python
#without increment
num=0  #initialisation
while num<5:
    print(f'the current num={num}')
    
```

as num isn't incremented so it remains less than 5 always so is printed for infinite times. 

```python
#with increment
num=0
while num<5:
    print(f'the current num={num}')
    num=num+1
```

```python
num=0
while num<10:
    print(num)
    num+=1
else:
    print('num is not less than 10.')
```

# 1. pass - does nothing at all

```python
#without pass
num=0
while num<5:
    
```

```python
#with pass
num=0
while num<5:
    pass
```

# 2.break - breaks out of the current closest enclosing loop

```python
#use of break
string='raseswari'
for letter in string:
    if letter=='w':
        break
    print(letter)
```

# 3.continue - goes to the top of closest enclosing loop

```python
#use of continue
string1='rajeswari'
for letter in string1:
    if letter=='a':
        continue
    print(letter)
```

```python
#example
'''Take integer inputs from user until he/she presses q ( Ask to press q to quit after every integer input ).
Print average and product of all numbers.'''
product=1
summation=0
count=0
loop_continue=True
while loop_continue:
    command=input('enter any number:').upper()
    count+=1
    if command=='Q':
        break
    else:
        print('press q to exit.')
        summation+=int(command)
        average=summation/count
        product*=int(command)
print(f'average={average}')
print(f'product={product}')
```

# Nested loop

```python
result=[]
for x in [2,4,6]:
    for y in [100,200,300]:
        result.append(x*y)
        print(result) #for knowing mechanism
print(f'\nfinal list={result}')
```
